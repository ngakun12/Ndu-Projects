```python
import statsmodels.api as sm

# Assuming 'df' is your DataFrame and 'name' is the qualitative variable to exclude
X = df.drop(columns=['mpg', 'name'])  # Predictor variables
y = df['mpg']  # Response variable

# Add a constant to the predictor variable set to represent the intercept
X = sm.add_constant(X)

# Handle missing values by dropping them (you may choose to fill them instead)
X = X.dropna()
y = y.loc[X.index]

# Fit the model
model = sm.OLS(y, X).fit()

# Print the summary
print(model.summary())
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Assuming `model` is the fitted OLS model from statsmodels
# 1. Residuals vs Fitted
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(model.fittedvalues, model.resid)
plt.xlabel('Fitted values')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted')
plt.axhline(y=0, color='grey', linestyle='dashed')

# 2. Normal Q-Q plot
plt.subplot(1, 2, 2)
sm.qqplot(model.resid, line='s')
plt.title('Normal Q-Q')

plt.tight_layout()
plt.show()

# 3. Scale-Location plot
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(model.fittedvalues, np.sqrt(np.abs(model.resid)))
plt.xlabel('Fitted values')
plt.ylabel('Sqrt(|Residuals|)')
plt.title('Scale-Location')

# 4. Residuals vs Leverage plot
plt.subplot(1, 2, 2)
sm.graphics.influence_plot(model, criterion="cooks")
plt.title('Residuals vs Leverage')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import statsmodels.formula.api as smf

# Assuming you've already set the correct path to your "carseats" CSV file
file_path = r"C:\Users\ngaku\Downloads\Carseats.csv"  # Update this path

# Load the data from the CSV file
df_carseats = pd.read_csv(file_path)

# Now, define the formula for your regression model
formula = 'Sales ~ Price + C(Urban) + C(US)'

# Fit the model
model = smf.ols(formula=formula, data=df_carseats).fit()

# Print the summary of the regression
print(model.summary())

```

```python
# Define the formula for the smaller model excluding 'Urban'
smaller_formula = 'Sales ~ Price + C(US)'

# Fit the smaller model
smaller_model = smf.ols(formula=smaller_formula, data=df_carseats).fit()

# Print the summary of the regression for the smaller model
print(smaller_model.summary())
```

```python
# Assuming 'smaller_model' is your fitted model object
conf_intervals = smaller_model.conf_int()
print(conf_intervals)
```

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Assuming 'smaller_model' is your fitted regression model

# Create an instance of influence measures
influence = smaller_model.get_influence()

# Obtain standardized residuals
standardized_residuals = influence.resid_studentized_internal

# Obtain leverage statistics
leverage = influence.hat_matrix_diag

# Plot standardized residuals vs. leverage
fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(leverage, standardized_residuals, alpha=0.5)
ax.set_xlabel('Leverage')
ax.set_ylabel('Standardized Residuals')
ax.set_title('Residuals vs. Leverage')
ax.axhline(y=0, color='r', linestyle='dashed')

# Adding Cook's distance contour lines to identify influential points
cooks_distances = influence.cooks_distance[0]
leverage_threshold = 3 * (smaller_model.df_model + 1) / len(smaller_model.model.endog)
ax.axvline(x=leverage_threshold, color='r', linestyle='dotted')
ax.annotate('Potential High Leverage Line', xy=(leverage_threshold, max(standardized_residuals)), color='red')

# Identify points with large Cook's distance
for i in range(len(cooks_distances)):
    if cooks_distances[i] > 4 / len(standardized_residuals):
        ax.annotate(i, (leverage[i], standardized_residuals[i]))

plt.show()
```

```python
import numpy as np
import statsmodels.api as sm

# Set seed for reproducibility
np.random.seed(123)

# Generate random data for Y
n = 100
Y = np.random.normal(size=n)

# Create X as a perfectly correlated version of Y, scaled by a factor
factor = 2
X = factor * Y

# Regression of Y onto X without an intercept
model_Y_on_X = sm.OLS(Y, X).fit()

# Regression of X onto Y without an intercept
model_X_on_Y = sm.OLS(X, Y).fit()

# Print the coefficients
print('Coefficient for regression Y onto X:', model_Y_on_X.params[0])
print('Coefficient for regression X onto Y:', model_X_on_Y.params[0])
```
